```python
import os

audio_dir = "../data/raw/audio"
files = os.listdir(audio_dir)
print(len(files))
```

```python
import librosa

durations = []
for f in files:
    y, sr = librosa.load(os.path.join(audio_dir, f), sr=16000)
    durations.append(len(y)/sr)

print(sum(durations)/len(durations))
```

```python
import matplotlib.pyplot as plt

plt.hist(durations)
plt.show()
```
